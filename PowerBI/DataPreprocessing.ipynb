#### Read multiple csv files

```python
import pandas as pd

file_paths = ["./Customers.csv",
              "./Products.csv",
              "./Sales_2015.csv",
              "./Sales_2016.csv",
              "./Sales_2017.csv"]

dataframe_names = ["Cus", "Products", "S2015", "S2016", "S2017"]

for i in range(len(file_paths)):
    globals()[dataframe_names[i]] = pd.read_csv(file_paths[i], encoding='ISO-8859-1')
```

```python
dataframes = [Cus, Products, S2015, S2016, S2017]
dataframe_names = ["Cus", "Products", "S2015", "S2016", "S2017"]
for df, name in zip(dataframes, dataframe_names):
    print("Data type of:", name)
    print(df.dtypes)
    print("*"*50)
```

#### Check the shape

```python
dataframes = [Cus, Products, S2015, S2016, S2017]
dataframe_names = ["Cus", "Products", "S2015", "S2016", "S2017"]
for df, name in zip(dataframes, dataframe_names):
    shape = df.shape
    print("The shape of:", name, "dataframe is", shape)
```

#### Check null values

```python
dataframes = [Cus, Products, S2015, S2016, S2017]
dataframe_names = ["Cus", "Products", "S2015", "S2016", "S2017"]
for df, name in zip(dataframes, dataframe_names):
    no_null = df.isnull().sum().sum()
    print("="*50)
    print("The dataframe:", name, "has", no_null, "null values")
    print(df.isnull().sum())
    print("="*50)
```

#### Check duplicated values

```python
dataframes = [Cus, Products, S2015, S2016, S2017]
dataframe_names = ["Cus", "Products", "S2015", "S2016", "S2017"]
for df, name in zip(dataframes, dataframe_names):
    dupl = df.duplicated().sum()
    print("The dataframe:", name, "has", dupl, "duplicated values")
```
